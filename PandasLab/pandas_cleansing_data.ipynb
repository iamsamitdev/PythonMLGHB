```python
# ลบรายการที่ซ้ำกันทั้งหมดออก
df.drop_duplicates(inplace=True)

df
```

#### สรุปการทำความสะอาดข้อมูลทุกเงื่อนไขที่ต้องการ

```python
df = pd.read_csv('datasets/exercise.csv')

# แทนที่ข้อมูลว่างด้วยคำสั่ง fillna
df.fillna(0, inplace=True)

# เปลี่ยนข้อมูลที่อยู่นอกช่วงให้ถูกต้อง
df.loc[2, 'Hight'] = 179.0

# ลบรายการที่ซ้ำกันทั้งหมดออก
df.drop_duplicates(inplace=True)

# ลบแถวที่มีรายการผิดรูปแบบออก
df.drop([3], inplace=True)

df
```

#### บันทึกข้อมูล DataFrame ที่ทำการ Clensing แล้ว เป็นไฟล์ CSV

```python
df.to_csv('datasets/exercise_clean.csv', index=False)
```
