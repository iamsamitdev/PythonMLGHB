#### 2. การวาดกราฟเส้น (Line chart)

```python
import matplotlib.pyplot as plt
import numpy as np

# กำหนดแกน x และ y
x = np.array([2018, 2019, 2020, 2021, 2022])
y = np.array([1050, 2452, 3456, 4653, 4052])

plt.plot(x,y)
plt.xticks(x)
plt.show()

plt.bar(x,y)
plt.xticks(x)
plt.show()
```

#### 3.การโหลดจาก csv มา plot กราฟ

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('datasets/gdp.csv')

# line plot
plt.plot(df['year'], df['gdp'])
plt.title('Thailand GDP')
plt.xlabel('Year')
plt.ylabel('GDP (Million Baht)')
plt.xticks(df['year'])
plt.show()
```

```python
# bar chart
plt.bar(
    df['year'], df['gdp'], 
    color=['green', 'red', 'blue', 'yellow', 'pink'], 
    edgecolor='black'
)
plt.title('Thailand GDP')
plt.xlabel('Year')
plt.ylabel('GDP (Million Baht)')
plt.xticks(df['year'])
plt.show()
```
