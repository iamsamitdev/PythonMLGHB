#### Step 2: อ่านข้อมูลภาพจากไฟล์

```python
import cv2

imagPath = "images/man-in-black.jpg"

img = cv2.imread(imagPath)

# ดูขนาดของภาพ
print(img.shape) # (height, width, channel) channel 3 = RGB
```

#### Step 3: Convert the Image to Grayscale แปลงภาพสีเป็นภาพขาวดำ

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(gray_img.shape)
```

#### Step 4: Load the Classifier โหลดตัวจำแนก

```python
face_classifier = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)
```

#### Step 5: Face Detection ตรวจจับใบหน้า

```python
face = face_classifier.detectMultiScale(
    gray_img, 
    scaleFactor=1.1,
    minNeighbors=10,
    minSize=(30, 30),
)
```

#### Step 6: Drawing a Bounding Box วาดกรอบสี่เหลี่ยมรอบใบหน้า

```python
for (x, y, w, h) in face:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 8) # สีเขียว ความหนา 8
```

#### Step 7: Displaying the Image แสดงภาพ

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
